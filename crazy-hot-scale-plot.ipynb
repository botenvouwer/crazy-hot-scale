```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')

crazy_hot_data = [
    ['Nora', 9.9, 3],
    ['Eline', 6.1, 8.4],
    ['De moeder van Jeroen', 9.5, 5.5]
]

df = pd.DataFrame(crazy_hot_data, columns=['who','crazy','hot'])

df.describe()
```

```python
df.head()
```

```python
fig = plt.figure()
ax = plt.axes()


plt.plot(np.arange(0,11,1), np.arange(0,11,1), c='r')
plt.plot([0,10], [4,4], c='b', linestyle='-.')

for index, row in df.iterrows():
    plt.scatter(row.hot, row.crazy, label=row.who, s=40, marker='*')

ax.legend()
plt.xticks(np.arange(0,11,1))
plt.yticks(np.arange(0,11,1))
plt.xlim(0, 10)
plt.ylim(0, 10);

plt.xlabel('Hot')
plt.ylabel('Crazy')

plt.show()
```
